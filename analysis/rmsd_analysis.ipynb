```python
import argparse
import matplotlib.pyplot as plt

def import_data(input_file):
    with open(input_file, 'r') as file:
        frames = []
        rmsds = []
        for line in file:
            if '#' not in line:
                pair = line.split()
                frames.append(float(pair[0]))
                rmsds.append(float(pair[1]))
        data = [frames, rmsds]
    return data

def trim_data(data, nth):
    trimmed_frames = []
    trimmed_rmsds = []
    for frame, rmsd in zip(data[0], data[1]):
        if frame % nth == 0:
            trimmed_frames.append(frame/100)
            trimmed_rmsds.append(rmsd)
    trimmed_data = [trimmed_frames, trimmed_rmsds]
    return(trimmed_data) 

def plot_data(all_data, ylim, title, labels, colors):
    fig, ax = plt.subplots()
    for data, label, color in zip(all_data, labels, colors):
        plt.plot(data[0],data[1], label=label, color=color)
    plt.title(title)
    plt.legend(loc=0, fontsize='small')
    plt.xlabel('Time(ns)')
    plt.ylabel('RMSD')
    plt.ylim(0, ylim)
    plt.margins(0)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    return fig
```

```python
bf = '/scratch/Users/shla9937/luger_lab/md_simulations/histone_folding/'
input_files = [bf+'0M_hpya_rep0'+'/analysis/rmsd_'+'0M_hpya'+'_300ns.dat',
               bf+'0M_hpya_rep1'+'/analysis/rmsd_'+'0M_hpya_rep1'+'_300ns.dat',
               bf+'0M_hpya_rep2'+'/analysis/rmsd_'+'0M_hpya_rep2'+'_300ns.dat',
               bf+'4M_hpya_rep0'+'/analysis/rmsd_'+'4M_hpya'+'_300ns.dat',
               bf+'4M_hpya_rep1'+'/analysis/rmsd_'+'4M_hpya_rep1'+'_300ns.dat',
               bf+'4M_hpya_rep2'+'/analysis/rmsd_'+'4M_hpya_rep2'+'_300ns.dat']

labels = []
labels = []
for file in input_files:
    file = file.split('/')
    labels.append(file[7])
colors = []
for label in labels:
    if '0M' in label:
        colors.append('grey')
    else:
        colors.append('orange')

nth = 100
ylim = 10
title = 'RMSD of hpya sims'

all_data = []
for file in input_files:
    data = import_data(file)
    trimmed_data = trim_data(data, nth)
    all_data.append(trimmed_data)
fig1 = plot_data(all_data, ylim, title, labels, colors)
```
